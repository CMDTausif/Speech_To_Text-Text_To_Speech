## NLP Project ----  Text_To_Speech

```python
from gtts import gTTS
from IPython.display import Audio

TTS1 = gTTS("A Hungry Stomach and An Empty Wallet Would teach You the Best Lesson Of Life")
TTS1.save("quotes1.wav")
sound_file1 = "quotes1.wav"
Audio(sound_file1, autoplay=False)
```

```python
TTS2 = gTTS("We've had cloning in the South for years. It's called cousins. ")
TTS2.save("quotes2.wav")
sound_file2 = "quotes2.wav"
Audio(sound_file2, autoplay=False)
```
